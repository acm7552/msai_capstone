```python
import os
import pandas as pd
import torch
import torchvision
from torchvision import datasets, transforms
import random

from ARIAdataset import buildARIA
from AiArtBench import buildArtBenchDataset as buildArtBench

import matplotlib.pyplot as plt
import torchvision.utils as vutils

from torch.utils.data import random_split, ConcatDataset, DataLoader

# simple helper notebook used to generate the figures in the README.
```

```python
dataloader = DataLoader(buildArtBench("test"), batch_size=128, shuffle=True)

images, labels = next(iter(dataloader))

# split classes
real_images = images[labels == 0][:25]
fake_images = images[labels == 1][:25]

# safety check
if len(real_images) < 1 or len(fake_images) < 1:
    print("One of the classes is missing in this batch.")
else:
    real_grid = vutils.make_grid(real_images, nrow=5)
    fake_grid = vutils.make_grid(fake_images, nrow=5)

    # plot side by side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(real_grid.permute(1, 2, 0))
    axes[0].axis('off')
    axes[0].set_title("Real")

    axes[1].imshow(fake_grid.permute(1, 2, 0))
    axes[1].axis('off')
    axes[1].set_title("AI")

    plt.tight_layout()
    plt.savefig("artbench.png")
    plt.show()
```
