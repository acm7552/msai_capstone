```python
# Helper notebook for generating a useful gitignore string.
import os
import random
```

```python
ROOT_DIR = "/home/stu14/s16/acm7552/Capstone/data"
NUM_SAMPLES = 1

IMAGE_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".gif", ".tiff", ".webp"}
```

```python
def generate_gitignore(root_dir, num_samples=10, seed=None):
    if seed is not None:
        random.seed(seed)

    all_images = []

    for dirpath, _, filenames in os.walk(root_dir):
        images = [
            f for f in filenames
            if os.path.splitext(f)[1].lower() in IMAGE_EXTS
        ]

        if not images:
            continue

        sample = random.sample(images, min(num_samples, len(images)))
        all_images.extend(sample)

    # builds gitignore string for images
    lines = []
    lines.append("data/**")
    lines.append("")
    lines.append("!data/**/")
    lines.append("")
    lines.append("# Provides one sample image for every category")

    for fname in all_images:
        lines.append(f"!**/{fname}")

    return "\n".join(lines)
```

```python
gitignore_str = generate_gitignore(ROOT_DIR, NUM_SAMPLES, seed=42)
print(gitignore_str)
```
